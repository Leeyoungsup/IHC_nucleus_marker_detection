```python
from glob import glob
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
image=Image.open("../../data/BCData/images/train/0.png")
positive_h5=(h5py.File('../../data/BCData/annotations/train/positive/0.h5','r'))
negative_h5=(h5py.File('../../data/BCData/annotations/train/negative/0.h5','r'))
positive_point=positive_h5['coordinates'][:]
negative_point=negative_h5['coordinates'][:]
```

```python
plt.imshow(image)
plt.scatter(positive_point[:,0],positive_point[:,1],c='r',s=5)
plt.scatter(negative_point[:,0],negative_point[:,1],c='b',s=5)
plt.show()
```
